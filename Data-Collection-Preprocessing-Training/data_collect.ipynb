```python
# Split category_list by '|', flatten the list, and get unique values
unique_categories = set(DF['category_list'].dropna().str.split('|').sum())

# Count the number of unique categories
num_unique_categories = len(unique_categories)

print(f"Total unique categories: {num_unique_categories}")
print(unique_categories)  # Display all unique category names
```

```python
# Filter rows where category_list does NOT contain '|'
df_single_category = DF[~DF['category_list'].str.contains('\|', na=False)]
category_counts = df_single_category['category_list'].value_counts()
```

```python
categories_to_keep = category_counts[category_counts >= 20].index

# Filter the DataFrame to include only those categories
df_filtered = df_single_category[df_single_category['category_list'].isin(categories_to_keep)]
```

```python
df_filtered['category_list'].value_counts()
```

```python
category_cols = {
    "is_software": "Software",
    "is_web": "Web",
    "is_mobile": "Mobile",
    "is_enterprise": "Enterprise",
    "is_advertising": "Advertising",
    "is_gamesvideo": "Games",
    "is_ecommerce": "E-Commerce",
    "is_biotech": "Biotech",
    "is_consulting": "Consulting"
}

data1["category_list"] = data1.apply(lambda row: "|".join([name for col, name in category_cols.items() if row[col] == 1]), axis=1)
```

```python
data1['category_list'].value_counts()
```

```python
data1 = data1[data1['category_list'].str.strip() != ""]  # Just in case there are empty strings

print(data1['category_list'].value_counts())
```

```python
# Convert dates from M/D/YYYY or MM/DD/YYYY to DD-MM-YYYY
def convert_to_ddmmyyyy(date_series):
    return pd.to_datetime(date_series, format='%m/%d/%Y', errors='coerce').dt.strftime('%Y-%m-%d')

# Apply to relevant columns
data1['founded_at'] = convert_to_ddmmyyyy(data1['founded_at'])
data1['first_funding_at'] = convert_to_ddmmyyyy(data1['first_funding_at'])
data1['last_funding_at'] = convert_to_ddmmyyyy(data1['last_funding_at'])

# Print result
print(data1.head())
```

```python
import pandas as pd

# Load the dataset
abcd = pd.read_csv('C:\\VS code\\FLEXI\\Projecty\\final_startup_dataset_v2.csv')

# Check the first few rows to identify the date columns
print(abcd[['founded_at', 'first_funding_at', 'last_funding_at']].head())

# Check if the dates follow the 'YYYYMMDD' format
def check_date_format(date_column):
    # Try matching the 'YYYYMMDD' format (8 digits with no separators)
    return abcd[date_column].apply(lambda x: bool(pd.to_datetime(str(x), format='%Y%m%d', errors='coerce')))

# Check the date columns
print(check_date_format('founded_at').value_counts())
print(check_date_format('first_funding_at').value_counts())
print(check_date_format('last_funding_at').value_counts())
```
